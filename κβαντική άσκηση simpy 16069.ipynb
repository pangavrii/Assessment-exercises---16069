## Μέρος Α — Κανονικοποίηση σωματιδίου σε άπειρο πηγάδι

Θεωρούμε το άπειρο πηγάδι στο διάστημα $0<x<L$ με κυματοσυνάρτηση

$$
\psi_n(x)=A\sin\left(\frac{n\pi x}{L}\right).
$$

Να βρείτε τη σταθερά κανονικοποίησης $A$ από τη σχέση

$$
\int_0^L |\psi_n(x)|^2 dx=1.
$$

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: χρησιμοποιήστε symbols(..., positive=True, integer=True), integrate, Eq, solve.

x, L, A = smp.symbols('x L A', positive = True)
n = smp.symbols('n', integer = True, positive = True)
psi = A * smp.sin(n * smp.pi * x / L)
norm_eq = smp.Eq(smp.integrate(psi**2, (x, 0, L)), 1)
sol_A = smp.solve(norm_eq, A)
display(sol_A[0])
```

## Μέρος Β — Αναμενόμενη τιμή της θέσης

Με τη θετική τιμή του $A$, υπολογίστε την αναμενόμενη τιμή

$$
\langle x\rangle = \int_0^L x |\psi_n(x)|^2 dx.
$$

Τι αποτέλεσμα περιμένετε λόγω συμμετρίας του πηγαδιού;

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: πάρτε A_pos = sqrt(2/L), ξαναορίστε την psi_norm και ολοκληρώστε

A_pos = smp.sqrt(2 / L)
psi_norm = A_pos * smp.sin(n * smp.pi * x / L)
x_mean = smp.integrate(x * psi_norm**2, (x, 0, L))
display(smp.simplify(x_mean))
```

## Μέρος Γ — Αναμενόμενη τιμή της ενέργειας με τελεστή

Ο τελεστής κινητικής ενέργειας σε μία διάσταση είναι

$$
\hat{T}=-\frac{\hbar^2}{2m}\frac{d^2}{dx^2}.
$$

Για το απειρόβαθο πηγάδι, το δυναμικό μέσα στο πηγάδι είναι μηδέν, άρα $\hat{H}=\hat{T}$. Υπολογίστε

$$
E_n = \int_0^L \psi_n(x)\,\hat{H}\psi_n(x)\,dx.
$$

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: ορίστε hbar, m ως θετικά symbols. Χρησιμοποιήστε diff(psi_norm, x, 2).

hbar, m = smp.symbols('hbar m', positive = True)
Hpsi = -(hbar**2 / (2 * m)) * smp.diff(psi_norm, x, 2)
E_n = smp.integrate(psi_norm * Hpsi * psi_norm, (x, 0, L))

display(smp.simplify(E_n))
```

## Μέρος Δ — Ορθογωνιότητα καταστάσεων

Ελέγξτε συμβολικά ότι οι δύο πρώτες ιδιοκαταστάσεις του άπειρου πηγαδιού είναι ορθογώνιες:

$$
\int_0^L \psi_1(x)\psi_2(x)dx=0.
$$

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: ορίστε psi1 και psi2 βάζοντας n=1 και n=2. Μετά κάντε integrate από 0 έως L.

psi1 = psi_norm.subs(n , 1)
psi2 = psi_norm.subs(n , 2)
overlap_12 = smp.integrate(psi1 * psi2, (x, 0, L))

display(smp.simplify(overlap_12))
```

## Μέρος Ε — Μεταθέτης θέσης και ορμής

Ο τελεστής ορμής είναι

$$
\hat{p}=-i\hbar\frac{d}{dx}.
$$

Για μία γενική συνάρτηση $f(x)$, υπολογίστε

$$
[x,\hat{p}]f(x)=x\hat{p}f(x)-\hat{p}(xf(x)).
$$

Το αποτέλεσμα πρέπει να οδηγήσει στη γνωστή σχέση $[x,\hat{p}]=i\hbar$.

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: χρησιμοποιήστε f = smp.Function('f')(x), smp.I και diff.

f = smp.Function('f')(x)
p_f = -smp.I * hbar * smp.diff(f , x)
p_xf = -smp.I * hbar * smp.diff(x * f , x)
comm_f = x * p_f * f - p_xf
display(smp.simplify(comm_f))
```

## Μέρος ΣΤ — Ανάπτυγμα μικρού χρόνου

Για μία στάσιμη κατάσταση, ο χρονικός παράγοντας είναι

$$
\exp\left(-\frac{iEt}{\hbar}\right).
$$

Βρείτε το ανάπτυγμα Taylor μέχρι και όρους τάξης $t^2$ γύρω από $t=0$.

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: χρησιμοποιήστε series(expr, t, 0, 3).

t, E = smp.symbols('t E', real = True)
time_factor = smp.exp(-smp.I * E * t/ hbar)
ser = smp.series(time_factor , t , 0 , 3)
display(ser)
```

## Μέρος Ζ — Άθροισμα πιθανοτήτων με `Sum`

Θεωρούμε συντελεστές ανάπτυξης

$$
c_n=\frac{\sqrt{6}}{\pi n}.
$$

Υπολογίστε συμβολικά

$$
\sum_{n=1}^{\infty} |c_n|^2.
$$

Χρησιμοποιήστε οπωσδήποτε `smp.Sum` και μετά `.doit()`.

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: χρησιμοποιήστε smp.Sum(c_n**2, (n, 1, smp.oo)).doit().

c_n = smp.sqrt(6) / (smp.pi * n)
prob_sum = smp.Sum(smp.Abs(c_n)**2, (n, 1, smp.oo)).doit()
prob_sum = smp.simplify(prob_sum)
display(prob_sum)
```

## Μέρος Η — Απλή διαφορική εξίσωση τύπου Schrödinger

Λύστε τη διαφορική εξίσωση

$$
\frac{d^2\psi}{dx^2}+k^2\psi=0.
$$

Αυτή εμφανίζεται στο ελεύθερο σωματίδιο ή μέσα στο άπειρο πηγάδι, πριν επιβληθούν συνοριακές συνθήκες.

```python
# Συμπληρώστε τον κώδικα.
# Υπόδειξη: χρησιμοποιήστε Function, Eq και dsolve.

k = smp.symbols('k', real = True, positive = True)
psi = smp.Function('psi')(x)
eq = smp.Eq(psi.diff(x, 2) + k**2 * psi, 0)
sol_ode = smp.dsolve(eq, psi)

display(sol_ode)
```
